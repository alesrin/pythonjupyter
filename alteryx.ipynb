```python
import numpy as np  # Librería para cálculos numéricos
import pandas as pd  # Librería para manipulación de datos tabulares
import matplotlib.pyplot as plt  # Librería para gráficos
from sklearn.linear_model import LinearRegression  # Modelo de regresión lineal de Scikit-learn

# 1. Generamos datos ficticios (Ingresos y Gastos en Restaurante)
np.random.seed(42)  # Fijamos una semilla para obtener siempre los mismos valores aleatorios
ingresos = np.random.randint(1000, 5000, 20)  # Genera 20 valores aleatorios entre 1000€ y 5000€
gastos = ingresos * 0.3 + np.random.randint(50, 300, 20)  # Relación ingresos-gastos con un poco de ruido

# 2. Creamos un DataFrame con los datos generados
df = pd.DataFrame({'Ingreso Mensual (€)': ingresos, 'Gasto Restaurante (€)': gastos})

# 3. Entrenamos el modelo de regresión lineal
X = df[['Ingreso Mensual (€)']]  # Variable independiente (predictora)
y = df['Gasto Restaurante (€)']  # Variable dependiente (respuesta)
modelo = LinearRegression()  # Creamos una instancia del modelo de regresión lineal
modelo.fit(X, y)  # Ajustamos el modelo a los datos (entrenamiento)

# 4. Hacemos una predicción: ¿Cuánto gastará alguien con un ingreso de 4000€?
nuevo_ingreso = np.array([[4000]])  # Creamos un array con el nuevo dato a predecir
gasto_predicho = modelo.predict(nuevo_ingreso)[0]  # Realizamos la predicción y extraemos el valor

# 5. Visualizamos los datos y la regresión lineal
plt.scatter(df['Ingreso Mensual (€)'], df['Gasto Restaurante (€)'], color='blue', label="Datos reales")  # Puntos reales
grafico_x = df['Ingreso Mensual (€)']  # Valores del eje X para la línea de regresión
grafico_y = modelo.predict(X)  # Predicciones del modelo
plt.plot(grafico_x, grafico_y, color='red', label="Línea de regresión")  # Línea de regresión
plt.scatter(4000, gasto_predicho, color='green', label=f"Predicción (4000€ → {gasto_predicho:.2f}€)", marker='o', s=100)  # Predicción específica
plt.xlabel("Ingreso Mensual (€)")  # Etiqueta del eje X
plt.ylabel("Gasto en Restaurante (€)")  # Etiqueta del eje Y
plt.title("Regresión Lineal: Ingresos vs. Gasto en Restaurante")  # Título del gráfico
plt.legend()  # Mostrar leyenda
plt.show()  # Mostrar gráfico

# 6. Mostramos la predicción
gasto_predicho  # Devuelve el valor predicho para 4000€
```
